```python
def create_labels(g):
    labels = {}
    for n in g.nodes():
        labels[n] = n
    return labels
arguments = {'node_size':1000, 'font_size':16, 'alpha':0.3}
```

```python
class Betweenness_Tuple:
    """
    Define tuple: (edge key, betweenness score)
    Useful for sorting
    """
    # useful to sort
    def __init__(self, key, betweenness):
        self.key = key
        self.betweenness = betweenness

    def __repr__(self):
        return repr((self.key, self.betweenness))

def girvan_newman(G, k):
    '''Return a list of k lists of nodes, each forming a 
    community.'''
    # we have a Graph with n components.
    # we find the edges with the most betweenness
    # we severe them and check the number of components
    # we check if k <= len(list(nx.connected_component_subgraphs(G)))

    if len(G.nodes()) == 1:
        return [G.nodes()]    
    
    def find_best_edge(G0):
        """
        Networkx implementation of edge_betweenness
        returns a dictionary. Make this into a list,
        sort it and return the edge with highest betweenness.
        """
        eb = nx.edge_betweenness_centrality(G0, normalized=False)
        eb_il = list(eb.items())
        eb_il.sort(key=lambda x: x[1], reverse=True)
        return eb_il[0]    

    def best_edge_across(connected_components):
        """
        input: list(nx.connected_component_subgraphs(G))
        output: edge with highest betweenness centrality
        """
        l = []
        for i in connected_components:
            be = find_best_edge(i)
            l.append(Betweenness_Tuple(be[0], be[1]))
        l = sorted(l, key=lambda x: x.betweenness, reverse=True)
        return l[0]

    def slicer(G):
        components = list(nx.connected_component_subgraphs(G))
        key_ = best_edge_across(components).key
        # try to use 
        #print(G.neighbors(key_[0]),G.neighbors(key_[1]))
        G.remove_edge(*key_)
        return G

    def slice_k_times(G, k):
        for i in range(k):
            slicer(G)
#        l = []
#        for i in list(nx.connected_component_subgraphs(G)):
#            l.append(i.nodes())
        return G #l
    
    return slice_k_times(G, k)

```

```python
# find node with largest degree
def grapher(G):
    node_and_degree = G.degree()
    (largest_hub, degree) = sorted(node_and_degree.items(), key = itemgetter(1))[-1]
    # Create ego graph of main hub
    hub_ego = nx.ego_graph(G, largest_hub)
    # Draw graph
    pos=nx.spring_layout(hub_ego)
    nx.draw(hub_ego, pos ,node_color='b',node_size=60,with_labels=True)
    # Draw ego as large and red
    nx.draw_networkx_nodes(hub_ego,pos,nodelist=[largest_hub],node_size=300,node_color='r')
    plt.savefig('ego_graph.png')
    plt.show()
```

def nx_girvan_newman(G):

    if len(G.nodes()) == 1:
        return [G.nodes()]

    def find_best_edge(G0):
        """
        Networkx implementation of edge_betweenness
        returns a dictionary. Make this into a list,
        sort it and return the edge with highest betweenness.
        """
        eb = nx.edge_betweenness_centrality(G0)
        eb_il = list(eb.items())
        eb_il.sort(key=lambda x: x[1], reverse=True)
        return eb_il[0][0]

    components = list(nx.connected_component_subgraphs(G))
    # removing the remaining edge with highest betweenness
    while len(components) == 1:
        G.remove_edge(*find_best_edge(G))
        components = list(nx.connected_component_subgraphs(G))

    result = [c.nodes() for c in components]

    for c in components:
        result.extend(girvan_newman(c))

    return result

```python
# PLEASE DO NOT MODIFY THE TEST CODE BELOW WHICH (ALONG WITH OTHER SUCH TESTS) 
# WILL BE USED TO TEST YOUR IMPLEMENTATION

class TestCommunities(unittest.TestCase):
    def setUp(self):
        
        self.g = nx.Graph()
        self.g.add_edges_from([
            (1, 2),(1, 3), (2, 3),
            (2, 4), (3, 4),
            (4, 5), (5, 6), (4, 6),
            (7, 8), (7, 9), (8, 9),
            (5, 9)
        ])

        
    def test_gn(self):    
        
        nx.draw(self.g, labels=create_labels(self.g), **arguments)
        
        components = girvan_newman(self.g, 2)
        c_set = {frozenset(c) for c in components}
        self.assertEqual(c_set,
            {
                frozenset([1, 2, 3, 4, 5, 6]),
                frozenset([7, 8, 9])
            })
    
        components = girvan_newman(self.g, 3)
        c_set = {frozenset(c) for c in components}
        self.assertEqual(c_set,
            {
                frozenset([1, 2, 3]),
                frozenset([4, 5, 6]),
                frozenset([7, 8, 9])
            })

suite = unittest.TestLoader().loadTestsFromModule(TestCommunities())
unittest.TextTestRunner().run(suite)
```
